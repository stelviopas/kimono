```r
if(white_noise != 0){
    gene_var <- mean(var(as.matrix(input_data[["gene"]])))
    input_data[["gene"]] <- input_data[["gene"]] + matrix(rnorm(nrow(input_data[["gene"]])*ncol(input_data[["gene"]]),mean = 0,sd = gene_var * white_noise),nrow = nrow(input_data[["gene"]]),ncol=ncol(input_data[["gene"]]))
    cnv_var <- mean(var(as.matrix(input_data[["cnv"]])))    
    input_data[["cnv"]] <- input_data[["cnv"]] + matrix(rnorm(nrow(input_data[["cnv"]])*ncol(input_data[["cnv"]]),mean = 0,sd = cnv_var * white_noise),nrow = nrow(input_data[["cnv"]]),ncol=ncol(input_data[["cnv"]]))
    beta_var <- mean(var(as.matrix(input_data[["methylation"]])))
    pre.methylation <- input_data[["methylation"]] + matrix(rnorm(nrow(input_data[["methylation"]])*ncol(input_data[["methylation"]]),mean = 0,sd = beta_var * white_noise),nrow = nrow(input_data[["methylation"]]),ncol=ncol(input_data[["methylation"]]))
    pre.methylation <- as.matrix(pre.methylation)
    pre.methylation[which(pre.methylation < 0)] <- 0
    pre.methylation[which(pre.methylation > 1)] <- 1
    input_data[['methylation']] <- pre.methylation
        
    
    input_data[["gene"]] <- setDT(input_data[["gene"]])
    input_data[["cnv"]] <- setDT(input_data[["cnv"]])
    input_data[['methylation']] <- setDT(as.data.frame(input_data[['methylation']]))
  }
```
